```python
import pandas as pd
```

```python
df_full=pd.read_csv("base_datos/poblacion.csv",header=2)
df_full
```

```python
df_full.info()
```

```python
df=df_full.drop(["Country Code","Indicator Name","Indicator Code","Unnamed: 65"],axis=1)
```

```python
df=pd.melt(df,id_vars=["Country Name"],var_name="year", value_name="pob")
```

```python
df["year"]=pd.Categorical(df["year"].apply(str))
df.dtypes
```

Se filtrara para algunos paises

```python
idx_filtro=df["Country Name"].isin(["Aruba","Colombia"]) 
idx_filtro
```

```python
df_sample=df[idx_filtro]
df_sample
```

# Crear multiples indices

```python
df_sample=df_sample.set_index(["Country Name","year"]).sort_index()
```

```python
df_sample.loc["Colombia":]
```

```python
df_sample.loc["Colombia",:].loc["2016",:]
```

```python
df_sample.xs(["Colombia","2016"])
```

Dado que hay dos incices que es el nombre y año entonces se puede hacer una busqueda solo en uno de ellos usando level

```python
df_sample.xs("2016",level="year")
```

```python
df_countries=df.set_index(["Country Name","year"]).sort_index(ascending=[True,True])
```

```python
ids=pd.IndexSlice
df_countries.loc[ids["Aruba":"Colombia","2015":"2017"],:].sort_index()
```

```python
df_countries.index.get_level_values(1)
```

Calcular la media de poblacion de cada año

```python
pd.options.display.float_format="{:.2f}".format
df_countries.mean(level="year")
```
